```python
import plotly_express as px
```

```python
stocks = px.data.stocks()
stocks.head()
```

```python
px.line(stocks, x="date", y=["GOOG", "AMZN", "MSFT"])
```

```python
fig = px.line(stocks.rename(columns=dict(GOOG="Google", AMZN="Amazon", MSFT="Microsoft")), 
        x="date", y=["Google", "Amazon", "Microsoft"], 
        template="simple_white", 
        labels=dict(variable="Company", date="Date", value="Value"),
        hover_name="variable", hover_data=dict(variable=False))

fig.update_layout(font_family="Rockwell", font_size=15,
                  legend=dict(orientation="h", title="", x=0.05, y=1.1))

fig.add_hline(y=1, line_width=2, line_dash="dot")
fig.show()
fig.write_html("../html/stocks.html")
# fig.write_image("../html/stocks.png")
```
